```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
    
# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
```

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None, parse_dates=True)
```

```python
properties.head()
```

```python
properties.info()
```

```python
df1 = properties.T
```

```python
df1.head()
df1.reset_index(inplace=True)
```

```python
df1.head()
```

```python
df1.columns = df1.iloc[0]
```

```python
df1.rename(columns={pd.NaT:'ID', 'Unnamed: 0': 'London_Borough'}, inplace=True)
```

```python
df1.drop(0, inplace=True)
```

```python
clean_properties = pd.melt(df1, id_vars= ['London_Borough', 'ID'])
```

```python
clean_properties.head()
```

```python
clean_properties.rename(columns={0:'Month', 'value':'average_price'}, inplace=True)
```

```python
clean_properties.dtypes
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
clean_properties['average_price'] = pd.to_numeric(clean_properties['average_price'])
```

```python
clean_properties.head()
```

```python
clean_properties.count()
```

```python
clean_properties.dropna(how='any', inplace=True)
```

```python
clean_properties
```

```python
filt = clean_properties['London_Borough'] == 'Barnet'
filt2 = clean_properties['London_Borough'] == 'Bexley'
clean_properties[filt2].plot(x='Month', y='average_price', color='red')
plt.title("Bexley Average Price Increase")
clean_properties[filt].plot(x='Month', y='average_price')
plt.title("Barnet Average Price Increase")
```

```python
clean_properties['Year'] = clean_properties['Month'].apply(lambda x: x.year)
```

```python
clean_properties.head()
```

```python
dfg = clean_properties.groupby(by=['London_Borough', 'Year']).mean()
```

```python
dfg.reset_index(inplace=True)
```

```python
dfg.head()
```

```python
input_filt = dfg[dfg['London_Borough'] == 'Camden']
first = input_filt[input_filt['Year'] == 1995]['average_price'].values
last = input_filt[input_filt['Year'] == 2020]['average_price'].values
print(first, last)
print((first/last) * 100)
```

```python
dfg.sort_values('London_Borough', inplace=True)
```

```python
def create_price_ratio(df2, filt):
    input_filt = df2[df2['London_Borough'] == filt]
    first = input_filt[input_filt['Year'] == 1995]['average_price'].values
    last = input_filt[input_filt['Year'] == 2020]['average_price'].values
    li = list((first/last) * 100)
    name_ratio = (filt, li[0])
    return name_ratio
    
```

```python
ratio_list = []
for burough in dfg['London_Borough']:
    ratio_list.append(create_price_ratio(dfg, burough))
```

```python
burough_ratios = set(ratio_list)
```

```python
burough_changes = pd.DataFrame(burough_ratios)
```

```python
burough_changes
```

```python
burough_changes.rename(columns={0: 'London_Borough', 1:'Price Change (%)'}, inplace=True)
```

```python
burough_changes.sort_values('Price Change (%)', ascending=False, inplace=True)
```

```python
burough_changes.head()
```

```python
plt.figure(figsize=(30,20))
sns.set_context('notebook', font_scale= 2)
ratio_chart = sns.barplot(x='London_Borough', y='Price Change (%)', data=burough_changes, palette='viridis')
ratio_chart.set_xticklabels(ratio_chart.get_xticklabels(), rotation=90)
plt.rcParams["axes.labelsize"] = 20
plt.xlabel('Borough')
plt.gcf().subplots_adjust(bottom=0.25)
plt.tight_layout
plt.savefig('Borough_Changes.png')
```
