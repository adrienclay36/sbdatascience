#### Import relevant packages:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import json
```

```python
login_count.head()
```

```python
login_count.isnull().sum()
```

No null values in the dataset

The above should represent the number of logins per 15 minute interval

```python
print(login_count.index.min())
print(login_count.index.max())
```

###### It looks like there are 9788 recorded logins between January 1st - April 13, 1970

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=login_count.index, y='login_count', data=login_count)
plt.xticks(rotation=90)
plt.tight_layout()
```

The above plot isn't horribly helpful, but does hint at seasonality. 

It might be helpful to calculate a moving average for a pretty significant window to get a better idea of how the number of logins are changing:

```python
login_rolling = login_count.rolling(2400).mean()
```

```python
plt.figure(figsize=(12,8))
plt.plot(login_rolling)
plt.xlabel('Login Time')
plt.ylabel('Login Count')
```

What we can see from this is that the average logins from January to April increased from approximately 7-8 on average to between 11 and 12 on average respectively.

Zooming in on a 30-day period will give us a better picture of when this up and down seasonality is happening:

```python
january = login_count.loc['1970-01-01':'1970-01-30']
february = login_count.loc['1970-02-01':'1970-02-28']
march = login_count.loc['1970-03-01':'1970-03-30']
april = login_count.loc['1970-04-01':'1970-04-13']
```

```python
months = [january, february, march, april]
month_names = ['January', 'February', 'March', 'April']
```

```python
for month in months:
    i = 0
    plt.figure(figsize=(12,8))
    sns.lineplot(x=month.index, y='login_count', data=month)
    plt.xlabel(month_names[i])
    plt.xticks(rotation=90)
    plt.tight_layout()
    i += 1
    print("\n")
```

# Conclusions:

1. The login counts appear to be increasing slightly as the weeks go by. It's hard to tell since we don't have the complete week of April, this would have allowed us to gain further insights to back that idea.

2. The peak days for logins are Saturday and Sunday (Weekends), and the peak times for logins are 12PM (Noon) but mostly the evenings, mostly from 10pm until 2am, but there are still logins occurring at a higher rate before dinner (when people are off work) and after dinner (when people are relaxing) vs. the morning time as a whole.

3. March was the peak month for logins, steadily increasing from January until then. Again, we can't know if they'd have increased more as we don't have the month of April in it's entirety. 

4. I would be interested to see what an ARIMA model would make of the rest of April, as it appears there is a slight drop in average logins RIGHT before we run out of data (as per the moving average calculation). This could really point to either another great increased, or maybe something has happened after that that caused the number of logins to drop.

# Model Building

### Pre-Processing

```python
df.head()
```

```python
df.drop(['last_trip_date', 'signup_date'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, drop_first=True, columns=['city'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('retained', axis=1)
y = df['retained']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## K-Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
kvalues = range(1,30)
```

```python
scores = []
for k in kvalues:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    
    scores.append(accuracy_score(y_test, pred))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(kvalues, scores, marker='.')
```

The best K-Value appears to be 20

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, plot_roc_curve
```

```python
print(classification_report(y_test, pred))
```

```python
plot_confusion_matrix(knn, X_test, y_test)
```

```python
plot_roc_curve(knn, X_test, y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
estimators = np.arange(1, 100)
```

```python
forest_scores = []
for n in estimators:
    forest = RandomForestClassifier(n_estimators=n)
    
    forest.fit(X_train, y_train)
    
    pred = forest.predict(X_test)
    
    forest_scores.append(accuracy_score(y_test, pred))
```

```python
plt.plot(estimators, forest_scores, marker='.')
```

```python
best_estimators = np.argmax(forest_scores)
```

```python
best_estimators
```

```python
forest = RandomForestClassifier(n_estimators=46)
forest.fit(X_train, y_train)
pred = forest.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
plot_roc_curve(forest, X_test, y_test)
```

Because of this, I would recommend the company focus on riders in the market who use the service throughout the weekdays, and take longer trips in general, as this service seems to suit them much better than those who do not have to go as far, and who do not need many rides during the weekdays.
